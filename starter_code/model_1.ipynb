# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.describe()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec',
                        'koi_period','koi_period_err1','koi_period_err2',
                        'koi_time0bk','koi_time0bk_err1','koi_time0bk_err2',
                        'koi_impact','koi_impact_err1','koi_impact_err2',
                        'koi_duration','koi_duration_err1','koi_duration_err2',
                        'koi_depth','koi_depth_err1','koi_depth_err2',
                        'koi_prad','koi_prad_err1','koi_prad_err2',
                        'koi_teq','koi_insol','koi_insol_err1','koi_insol_err2',
                        'koi_model_snr','koi_steff','koi_steff_err1','koi_steff_err2',
                        'koi_slogg','koi_slogg_err1','koi_slogg_err2',
                        'koi_srad','koi_srad_err1','koi_srad_err2',
                        'ra','dec','koi_kepmag']]
selected_features.head(10)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# get X and y variables
y = df["koi_disposition"]
X = selected_features

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

print("Distribution of classes within data")
print("_"*30)
print(pd.Series(y).value_counts())

print("Distribution of classes within training set")
print("_"*30)
print(pd.Series(y_train).value_counts())
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Import dependencies
from sklearn.preprocessing import LabelEncoder, MinMaxScaler,StandardScaler
from tensorflow.keras.utils import to_categorical

# scale the data
X_scaler =  StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)


print(encoded_y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

# Train the Model



```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

GridSearch requires 2 things:

A model
A dictionary of the hyperparameters and the associated lists of values to test with.
In the example below, we're tuning 2 hyperparameters of an SVC model:

C is the penalty term for SVMs. A higher value means the model will be more strict about preventing points from crossing the boundary. A lower value will soften the margins and allow for some leniency with points crossing over.
gamma defines how far the influence of single training example reaches. A high gamma means our decision boundary is more influenced by closeby points. A low gamma increases the influence of points further away from the decision boundary.

```python
# Train the model with GridSearch
param_grid = {'C': [0.001, 0.01, 1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.05, 0.5]}

grid = GridSearchCV(model, param_grid, verbose=3, scoring='f1')
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
